```python
import pandas as pd

# ---------------------------------------------------------
# 1. LOAD THE DATASET
# ---------------------------------------------------------

data = pd.read_csv(
    "automobile_service_descriptive_analytics_2000_records.csv"
)

# Convert required columns to numeric format
numeric_columns = [
    "total_bill_inr",
    "turnaround_hours",
    "customer_rating"
]

for column in numeric_columns:
    data[column] = pd.to_numeric(
        data[column],
        errors="coerce"
    )

# Standardize Yes/No values
data["repeat_visit_30d"] = (
    data["repeat_visit_30d"]
    .astype(str)
    .str.strip()
    .str.title()
)

# Remove rows missing important values
data = data.dropna(
    subset=[
        "service_id",
        "total_bill_inr",
        "turnaround_hours",
        "customer_rating"
    ]
)
print(data)
```

```python
# Total jobs
total_jobs = data["service_id"].count()
print(f"Total jobs: {total_jobs}")
#complete the rest
```

```python
total_revenue = data["total_bill_inr"].sum()
print(f"Total revenue: INR {total_revenue:,.0f}")
```

```python
average_bill = total_revenue / total_jobs
print(f"Average bill: INR {average_bill:,.2f}")
```

```python
median_bill = data["total_bill_inr"].median()
print(f"Median bill: INR {median_bill:,.0f}")
```

```python
average_tat = data["turnaround_hours"].mean()
print(f"Average TAT: {average_tat:.2f} hours")
```

```python
median_tat = data["turnaround_hours"].median()
print(f"Median TAT: {median_tat:.2f} hours")
```

```python
average_rating = data["customer_rating"].mean()
print(f"Average rating: {average_rating:.2f} / 5")
```

```python
repeat_visits_count = data[data["repeat_visit_30d"] == "Yes"].shape[0]
repeat_visits_percentage = (repeat_visits_count / total_jobs) * 100
print(f"Repeat visits: {repeat_visits_count} ({repeat_visits_percentage:.1f}%)")
```
